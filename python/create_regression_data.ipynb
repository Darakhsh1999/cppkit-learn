```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression, load_diabetes
from sklearn.model_selection import train_test_split
```

## Diabetes Data

```python
data = load_diabetes()
```

## Custom Regression Data

```python
X, y = make_regression(
    n_samples=1000,
    n_features=20,
    n_informative=15,
    n_targets=1,
    noise=0.5
)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1999)
```

```python
df_train = pd.DataFrame(
    np.column_stack((X_train, y_train)),
    columns=[f'feature_{i}' for i in range(X_train.shape[1])] + ['target']
)
df_train.head()
```

```python
df_test = pd.DataFrame(
    np.column_stack((X_test, y_test)),
    columns=[f'feature_{i}' for i in range(X_test.shape[1])] + ['target']
)
df_test.head()
```

```python
df_train.to_csv("regression_data_train.csv", index=False)
df_test.to_csv("regression_data_test.csv", index=False)
```
