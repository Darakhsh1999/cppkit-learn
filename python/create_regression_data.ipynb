```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_regression, load_diabetes
```

## Diabetes Data

```python
data = load_diabetes()
```

## Custom Regression Data

```python
X, y = make_regression(
    n_samples=1000,
    n_features=20,
    n_informative=15,
    n_targets=1,
    noise=0.5
)
```

```python
df = pd.DataFrame(
    np.column_stack((X, y)),
    columns=[f'feature_{i}' for i in range(X.shape[1])] + ['y']
)
df.head()
```

```python
df.to_csv("regression_data.csv", index=False)
```
