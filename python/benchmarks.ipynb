## Linear Regression

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data ---
algorithms = [
    "Data Loading",
    "Linear Regression",
]

# Time in milliseconds (replace with your actual measurements)
lib_a = [2.063,0.103]
lib_b = [4.0734,1.2999]

lib_a_label = "Cppkit-learn"
lib_b_label = "Scikit-learn"

# --- Layout ---
n = len(algorithms)
bar_height = 0.35
y = np.arange(n)

fig, ax = plt.subplots(figsize=(10, 7))

bars_a = ax.barh(y + bar_height / 2, lib_a, height=bar_height,
                 label=lib_a_label, color="#4C72B0", edgecolor="white", linewidth=0.5)
bars_b = ax.barh(y - bar_height / 2, lib_b, height=bar_height,
                 label=lib_b_label, color="#DD8452", edgecolor="white", linewidth=0.5)

# --- Value labels on bars ---
for bar in bars_a:
    width = bar.get_width()
    ax.text(width + 1.5, bar.get_y() + bar.get_height() / 2,
            f"{width:.1f}", va="center", ha="left", fontsize=8, color="#333333")

for bar in bars_b:
    width = bar.get_width()
    ax.text(width + 1.5, bar.get_y() + bar.get_height() / 2,
            f"{width:.1f}", va="center", ha="left", fontsize=8, color="#333333")

# --- Axes & labels ---
ax.set_yticks(y)
ax.set_yticklabels(algorithms, fontsize=11)
ax.set_xlabel("Time (ms)", fontsize=11)
ax.set_title(f"Algorithm Performance: {lib_a_label} vs {lib_b_label}",
             fontsize=13, fontweight="bold", pad=14)

# Extend x-axis so value labels don't get clipped
max_val = max(max(lib_a), max(lib_b))
ax.set_xlim(0, max_val * 1.18)

# --- Polish ---
ax.legend(fontsize=10, framealpha=0.9)
ax.spines[["top", "right"]].set_visible(False)
ax.xaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)

plt.tight_layout()
plt.savefig("grouped_bar_chart.png", dpi=150, bbox_inches="tight")
plt.show()
```
